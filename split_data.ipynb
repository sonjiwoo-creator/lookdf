```python
import shutil
from pathlib import Path

# data 폴더 생성
Path("./data").mkdir(exist_ok=True)

# 작은 파일들은 이름만 깔끔하게 바꿔서 복사
rename_map = {
    'orders_202604091310.csv':               'orders.csv',
    'order_items_202604091310.csv':          'order_items.csv',
    'products_202604091309.csv':             'products.csv',
    'users_202604091308.csv':                'users.csv',
    'distribution_centers_202604091311.csv': 'distribution_centers.csv',
}

for old, new in rename_map.items():
    shutil.copy(old, f'./data/{new}')
    print(f"✅ {new}")
```

```python
import pandas as pd

events = pd.read_csv("./events_202604091311.csv")
print(f"원본: {len(events):,} 행")

n_splits = 6
chunk_size = len(events) // n_splits + 1

for i in range(n_splits):
    start = i * chunk_size
    end   = start + chunk_size
    chunk = events.iloc[start:end]
    chunk.to_csv(f"./data/events_part{i+1}.csv", index=False)
    print(f"✅ events_part{i+1}.csv: {len(chunk):,} 행")

del events
```

```python
inventory_items = pd.read_csv("./inventory_items_202604091311.csv")
print(f"원본: {len(inventory_items):,} 행")

n_splits = 2
chunk_size = len(inventory_items) // n_splits + 1

for i in range(n_splits):
    start = i * chunk_size
    end   = start + chunk_size
    chunk = inventory_items.iloc[start:end]
    chunk.to_csv(f"./data/inventory_items_part{i+1}.csv", index=False)
    print(f"✅ inventory_items_part{i+1}.csv: {len(chunk):,} 행")

del inventory_items
```
