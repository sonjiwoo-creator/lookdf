```python
# ============================================
# users 데이터 전처리
# ============================================

import pandas as pd

# low_memory=False : 컬럼 데이터 타입을 전체 파일 읽은 후 결정
# → postal_code 같이 숫자/문자 섞인 컬럼 경고 방지
df_users = pd.read_csv('../data/users_202604091308.csv', 
                        low_memory=False)

# 기본 정보 확인
print("=== users 기본 정보 ===")
print("shape:", df_users.shape)
print(df_users.head())
```

```python
# -----------------------------------------------
# 중복값 확인 및 제거
# -----------------------------------------------

# 중복 행이 몇 개인지 확인
print("중복 행 수:", df_users.duplicated().sum())

# 중복 행 제거
# keep='first' : 중복 중 첫 번째 행만 남기고 나머지 제거
df_users = df_users.drop_duplicates(keep='first')

# 제거 후 shape 확인
print("중복 제거 후 shape:", df_users.shape)
```

```python
# -----------------------------------------------
# 결측치 확인 및 처리
# -----------------------------------------------

# 각 컬럼별 결측치 개수 확인
print("결측치 확인:")
print(df_users.isnull().sum())

# city 컬럼 결측치 → 'Unknown'으로 채우기
# fillna() : 결측치를 특정 값으로 채우는 함수
print("\n처리 전 city 결측치:", df_users['city'].isnull().sum())
df_users['city'] = df_users['city'].fillna('Unknown')
print("처리 후 city 결측치:", df_users['city'].isnull().sum())
```

```python
# -----------------------------------------------
# 최종 확인
# -----------------------------------------------

# 전처리 완료 후 최종 shape 및 결측치 확인
print("최종 shape:", df_users.shape)
print("\n최종 결측치 확인:")
print(df_users.isnull().sum())
```
