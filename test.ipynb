```python
import pandas as pd 

orders = pd.read_csv("./orders_202604091310.csv")
distribution_centers = pd.read_csv("./distribution_centers_202604091311.csv")
events = pd.read_csv("./events_202604091311.csv")
inventory_items = pd.read_csv("./inventory_items_202604091311.csv")
order_items = pd.read_csv("./order_items_202604091310.csv")
products = pd.read_csv("./products_202604091309.csv")
users = pd.read_csv("./users_202604091308.csv")
```

## 날짜 컬럼 변경

```python
# orders: 날짜 관련 컬럼 4개 → datetime으로 변환, status/gender → category타입으로 변환
orders['created_at']   = pd.to_datetime(orders['created_at'],   errors='coerce', utc=True)
orders['shipped_at']   = pd.to_datetime(orders['shipped_at'],   errors='coerce', utc=True)
orders['delivered_at'] = pd.to_datetime(orders['delivered_at'], errors='coerce', utc=True)
orders['returned_at']  = pd.to_datetime(orders['returned_at'],  errors='coerce', utc=True)

orders['status'] = orders['status'].astype('category')
orders['gender'] = orders['gender'].astype('category')

orders.dtypes
```

```python
#UTC기준으로 변환했기에 뒤에 +00:00을 남겨둬진것이고, 혹시 한국 시간으로 변환해야할수도 있을거같아서 그냥 냅둔 상태입니다. 필요하다면 dt.tz_convert('Asia/Seoul')이나 다른 도시로 변환하시면 될 것 같습니다.
orders['created_at'].info
```

```python
#이런식으로 한국 시간으로 변환할 수도 있음
orders['created_at'].dt.tz_convert('Asia/Seoul')
```

```python
# order_items: 날짜 관련 컬럼 4개 → datetime으로 변환, status → category타입으로 변환, sale_price → 반올림하고 int형식으로 변환
order_items['created_at']   = pd.to_datetime(order_items['created_at'],   errors='coerce', utc=True)
order_items['shipped_at']   = pd.to_datetime(order_items['shipped_at'],   errors='coerce', utc=True)
order_items['delivered_at'] = pd.to_datetime(order_items['delivered_at'], errors='coerce', utc=True)
order_items['returned_at']  = pd.to_datetime(order_items['returned_at'],  errors='coerce', utc=True)

order_items['status'] = order_items['status'].astype('category')
order_items['sale_price'] = order_items['sale_price'].round().astype('int64')

order_items.dtypes
```

```python
# inventory_items: 날짜 관련 컬럼 2개 → datetime으로 변환, 범주형 3개 → category타입으로 변환, 가격 2개 → 반올림하고 int형식으로 변환
inventory_items['created_at'] = pd.to_datetime(inventory_items['created_at'], errors='coerce', utc=True)
inventory_items['sold_at']    = pd.to_datetime(inventory_items['sold_at'],    errors='coerce', utc=True)

inventory_items['product_category']   = inventory_items['product_category'].astype('category')
inventory_items['product_brand']      = inventory_items['product_brand'].astype('category')
inventory_items['product_department'] = inventory_items['product_department'].astype('category')

inventory_items['cost']                 = inventory_items['cost'].round().astype('int64')
inventory_items['product_retail_price'] = inventory_items['product_retail_price'].round().astype('int64')

inventory_items.dtypes
```

```python
# products: 범주형 컬럼 3개 → category타입으로 변환, 가격 2개 → 반올림하고 int형식으로 변환
products['category']   = products['category'].astype('category')
products['brand']      = products['brand'].astype('category')
products['department'] = products['department'].astype('category')

products['cost']         = products['cost'].round().astype('int64')
products['retail_price'] = products['retail_price'].round().astype('int64')

products.dtypes
```

```python
# users: created_at → datetime으로 변환, 범주형 3개 → category타입으로 변환, postal_code → str타입으로 변환
users['created_at'] = pd.to_datetime(users['created_at'], errors='coerce', utc=True)

users['gender']         = users['gender'].astype('category')
users['country']        = users['country'].astype('category')
users['traffic_source'] = users['traffic_source'].astype('category')

users['postal_code'] = users['postal_code'].astype('str')

users.dtypes
```

```python
# events: created_at → datetime으로 변환, 범주형 3개 → category타입으로 변환
# user_id는 정연님이 처리할 예정이므로 기존 형식인 float64 유지하였음
events['created_at'] = pd.to_datetime(events['created_at'], errors='coerce', utc=True)

events['browser']        = events['browser'].astype('category')
events['traffic_source'] = events['traffic_source'].astype('category')
events['event_type']     = events['event_type'].astype('category')

events.dtypes
```

```python
# ============================================
# users 데이터 전처리
# ============================================

import pandas as pd

# low_memory=False : 컬럼 데이터 타입을 전체 파일 읽은 후 결정
# → postal_code 같이 숫자/문자 섞인 컬럼 경고 방지
df_users = pd.read_csv('users_202604091308.csv', 
                        low_memory=False)

# 기본 정보 확인
print("=== users 기본 정보 ===")
print("shape:", df_users.shape)
print(df_users.head())
```

```python
# -----------------------------------------------
# 중복값 확인 및 제거
# -----------------------------------------------

# 중복 행이 몇 개인지 확인
print("중복 행 수:", df_users.duplicated().sum())

# 중복 행 제거
# keep='first' : 중복 중 첫 번째 행만 남기고 나머지 제거
df_users = df_users.drop_duplicates(keep='first')

# 제거 후 shape 확인
print("중복 제거 후 shape:", df_users.shape)
```

```python
# -----------------------------------------------
# 결측치 확인 및 처리
# -----------------------------------------------

# 각 컬럼별 결측치 개수 확인
print("결측치 확인:")
print(df_users.isnull().sum())

# city 컬럼 결측치 → 'Unknown'으로 채우기
# fillna() : 결측치를 특정 값으로 채우는 함수
print("\n처리 전 city 결측치:", df_users['city'].isnull().sum())
df_users['city'] = df_users['city'].fillna('Unknown')
print("처리 후 city 결측치:", df_users['city'].isnull().sum())
```

```python
# -----------------------------------------------
# 최종 확인
# -----------------------------------------------

# 전처리 완료 후 최종 shape 및 결측치 확인
print("최종 shape:", df_users.shape)
print("\n최종 결측치 확인:")
print(df_users.isnull().sum())
```
